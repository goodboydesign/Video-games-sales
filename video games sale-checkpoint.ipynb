```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("vgchartz-2024.csv")

# Inspect the first few rows
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Handle missing values
df['critic_score'].fillna(df['critic_score'].mean(), inplace=True)
df['total_sales'].fillna(0, inplace=True)
df['na_sales'].fillna(0, inplace=True)
df['jp_sales'].fillna(0, inplace=True)
df['pal_sales'].fillna(0, inplace=True)
df['other_sales'].fillna(0, inplace=True)


# Remove duplicates
df.drop_duplicates(inplace=True)
df.head()
```

```python
# Correcting placeholders
df['img'].replace('/games/boxart/default.jpg', np.nan, inplace=True)
df['publisher'].replace('Unknown', np.nan, inplace=True)
df['developer'].replace('Unknown', np.nan, inplace=True)
df.head()
```

```python
# Descriptive statistics
print("Descriptive Statistics:")
print(df.describe())
```

```python
# Histograms for numerical columns
numerical_columns = ['critic_score', 'total_sales', 'na_sales', 'jp_sales', 'pal_sales', 'other_sales']
df[numerical_columns].hist(bins=20, figsize=(14, 10), layout=(3, 2))
plt.suptitle('Histograms of Numerical Columns')
plt.show()
```

```python
# Box plots for numerical columns
df[numerical_columns].plot(kind='box', subplots=True, layout=(3, 2), figsize=(14, 10), title='Box plots of Numerical Columns')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df[numerical_columns].corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
# Detecting anomalies using Z-score
from scipy.stats import zscore
```

```python
z_scores = np.abs(zscore(df[numerical_columns]))
anomalies = df[(z_scores > 3).any(axis=1)]
print("Anomalies detected:")
print(anomalies)
```

```python
import plotly.express as px
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
```

```python
# Initialize the Dash app
app = dash.Dash(__name__)
```

```python
# App layout
app.layout = html.Div([
    html.H1("VGChartz 2024 Data Dashboard"),dcc.Dropdown(
        id='genre-dropdown',
        options=[{'label': genre, 'value': genre} for genre in df['genre'].unique()],
        value='Action',
        clearable=False,
        style={'width': '50%'}
    ),

    dcc.Graph(id='sales-by-genre'),
    dcc.Graph(id='critic-score-distribution'),
    dcc.Graph(id='sales-correlation-heatmap')
])
```

```python
# Callback to update graphs based on selected genre
@app.callback(
    [Output('sales-by-genre', 'figure'),
     Output('critic-score-distribution', 'figure'),
     Output('sales-correlation-heatmap', 'figure')],
    [Input('genre-dropdown', 'value')])
def update_graphs(selected_genre):
    filtered_df = df[df['genre'] == selected_genre]
    # Sales by genre
    sales_fig = px.bar(filtered_df, x='title', y='total_sales', color='console', title=f'Total Sales for {selected_genre} Games')
# Critic score distribution
    critic_fig = px.histogram(filtered_df, x='critic_score', nbins=20, title=f'Critic Score Distribution for {selected_genre} Games')
    return sales_fig, critic_fig, heatmap_fig
```

```python
# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
    
```
